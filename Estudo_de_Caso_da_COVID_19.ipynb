Embora seja tratada como uma novidade, esta não é a primeira vez que ocorre uma pandemia na história da humanidade, houve outros surtos de doenças que causaram enormes danos ao longo das eras, como por exemplo: 
* A Varíola, que assolou o mundo durante vários séculos, onde foram identificadas as primeiras evidências em múmias datadas do século III e sendo erradicada em 1980 de acordo com a Organização Mundial de Saúde(OMS).

* A Gripe Espanhola, onde se estima que entre 40 milhões e 50 milhões de pessoas tenham morrido em 1918 decorrente de seus sintomas.
* A Gripe Suína, sendo considerada a primeira pandemia do século XXI, matando 16 mil pessoas ao redor do mundo em 2009.

Apesar do mundo globalizado influenciar na disseminação de doenças infecciosas, ele também traz o benefício da conexão humana, onde as informações e a ajuda são obtidas ou enviadas em pouquíssimo tempo para qualquer canto do mundo, o que não era possível antigamente.

##### Calculando o IQR Score das variáveis

Para fazer a limpeza dos outliers, vamos utilizar a regra do IQR Score para termos os limites superior e inferior de cada viariável. 

Como o IQR será utilizado para todas as variáveis, vamos transformar em uma função.



```
# Definindo uma função para calcular o IQR das variáveis
def iqr(x):
  # Calculando o IQR Score
  q1 = x.quantile(.25) # Primeiro Quartil(25%)
  q3 = x.quantile(.75) # Terceiro Quartil(75%)
  IQR_Score = q3 - q1

  # Definindo os limites superior e inferior
  superior = q3 + 1.5 * IQR_Score
  inferior = q1 - 1.5 * IQR_Score

  print("O IQR Score da variável {} é {}".format(x.name,IQR_Score))
  print("O limite superior é de {}.".format(superior))
  print("O limite inferior é de {}.".format(inferior))
  return
```

Agora com a função pronta, pode-mos analisar cada variável numérica e ter os limites:

```
# Criando uma lista com as variáveis que serão analisadas
lista_variaveis_iqr = [df_data_clean.total_cases,df_data_clean.new_cases,df_data_clean.new_cases_smoothed,
                       df_data_clean.total_deaths,df_data_clean.new_deaths,df_data_clean.new_deaths_smoothed,
                       df_data_clean.total_cases_per_million,df_data_clean.new_cases_per_million,df_data_clean.new_cases_smoothed_per_million,
                       df_data_clean.total_deaths_per_million,df_data_clean.new_deaths_per_million,df_data_clean.new_deaths_smoothed_per_million,
                       df_data_clean.reproduction_rate,df_data_clean.new_tests,df_data_clean.total_tests,df_data_clean.total_tests_per_thousand,
                       df_data_clean.new_tests_per_thousand,df_data_clean.new_tests_smoothed,df_data_clean.new_tests_smoothed_per_thousand,
                       df_data_clean.positive_rate,df_data_clean.tests_per_case,df_data_clean.stringency_index,df_data_clean.population,
                       df_data_clean.population_density,df_data_clean.median_age,df_data_clean.aged_65_older,df_data_clean.aged_70_older,
                       df_data_clean.gdp_per_capita,df_data_clean.extreme_poverty,df_data_clean.cardiovasc_death_rate,
                       df_data_clean.diabetes_prevalence,df_data_clean.female_smokers,df_data_clean.male_smokers,df_data_clean.handwashing_facilities,
                       df_data_clean.hospital_beds_per_thousand,df_data_clean.life_expectancy,df_data_clean.human_development_index]

# Criando um loop para executar a função iqr para cada variável na lista
for i in lista_variaveis_iqr:
  iqr(i)
  print('') #espaçamento de linha entre cada chamada da função
```

##### Removendo as outliers

Com a informação dos limites obtidos acima, removeremos os valores que estão abaixo de 0 para o limite inferior e os valores que estão acima do limite superior, porém temos que nos atentar com o fato de que o banco de dados é atualizado com novas informações constantemente, assim causando alterações no IQR_Score e consequantemente nos limites obtidos.

Para contornar esse problema de valores desatualizados, podemos criar uma nova função para que o limite superior acompanhe as mudanças em IQR_Score, sempre que o banco de dados for atualizado.

```
# Definindo a função para o limite superior
def limite_sup(x):
  # Calculando o IQR Score
  q1 = x.quantile(.25) # Primeiro Quartil(25%)
  q3 = x.quantile(.75) # Terceiro Quartil(75%)
  IQR_Score = q3 - q1

  # Definindo os limites superior
  superior = q3 + 1.5 * IQR_Score
 
  return superior
```

Como faremos o processo para várias variáveis, podemos criar mais uma função de remoção.

Nesta função de limpeza temos que a remoção das outliers superiores acompanhará a atualização do banco de dados e todos os valores negativos serão removidos. 

```
# Clonando a variável antes da remoção das outliers
df_data_clean_outlier = df_data_clean.copy()

# Definindo a função para remover as outliers
def limpeza_outlier(x):
  a = df_data_clean_outlier
  a.drop(a[x > limite_sup(x)].index, axis=0, inplace= True)
  a.drop(a[x < 0].index, axis=0, inplace= True)
  return

# Definindo a lista das variáveis para a limpeza
lista_limpeza_outlier = [df_data_clean_outlier.total_cases,df_data_clean_outlier.new_cases,df_data_clean_outlier.new_cases_smoothed,
                         df_data_clean_outlier.total_deaths,df_data_clean_outlier.new_deaths,df_data_clean_outlier.new_deaths_smoothed,
                         df_data_clean_outlier.total_cases_per_million,df_data_clean_outlier.new_cases_per_million,df_data_clean_outlier.new_cases_smoothed_per_million,
                         df_data_clean_outlier.total_deaths_per_million,df_data_clean_outlier.new_deaths_per_million,df_data_clean_outlier.new_deaths_smoothed_per_million,
                         df_data_clean_outlier.reproduction_rate,df_data_clean_outlier.new_tests,df_data_clean_outlier.total_tests,df_data_clean_outlier.total_tests_per_thousand,
                         df_data_clean_outlier.new_tests_per_thousand,df_data_clean_outlier.new_tests_smoothed,df_data_clean_outlier.new_tests_smoothed_per_thousand,
                         df_data_clean_outlier.positive_rate,df_data_clean_outlier.tests_per_case,df_data_clean_outlier.stringency_index,df_data_clean_outlier.population,
                         df_data_clean_outlier.population_density,df_data_clean_outlier.median_age,df_data_clean_outlier.aged_65_older,df_data_clean_outlier.aged_70_older,
                         df_data_clean_outlier.gdp_per_capita,df_data_clean_outlier.extreme_poverty,df_data_clean_outlier.cardiovasc_death_rate,
                         df_data_clean_outlier.diabetes_prevalence,df_data_clean_outlier.female_smokers,df_data_clean_outlier.male_smokers,df_data_clean_outlier.handwashing_facilities,
                         df_data_clean_outlier.hospital_beds_per_thousand,df_data_clean_outlier.life_expectancy,df_data_clean_outlier.human_development_index]

# Criando um loop para executar a função limpeza_outlier
for i in lista_limpeza_outlier:
  limpeza_outlier(i)
```

Vamos destacar os 4 países obtidos pelos DataFrames em gráficos para podermos visualizar melhor suas proporções e fazermos comparações: 

```
# Localizando os dados sobre os Estados Unidos, Brasil, Índia e México
usa = df_data_clean.loc[df_data_clean.iso_code == 'USA']
bra = df_data_clean.loc[df_data_clean.iso_code == 'BRA']
ind = df_data_clean.loc[df_data_clean.iso_code == 'IND']
mex = df_data_clean.loc[df_data_clean.iso_code == 'MEX']

# Criando um novo DataFrame com os países selecionados para casos e óbitos
paises_selecionados = usa.append(bra).append(ind).append(mex)

# Localizando as variáveis 'total_cases' e 'total_deaths' na data mais recente
total_casos_data_recente = paises_selecionados.loc[paises_selecionados.date == data_recente,['date','location','total_cases']].sort_values(by="total_cases", ascending=False)
total_mortes_data_recente = paises_selecionados.loc[paises_selecionados.date == data_recente,['date','location','total_deaths']].sort_values(by="total_deaths", ascending=False)

# Plotando os gráficos de barra para as duas variáveis
sns.set_theme()
fig, ax = plt.subplots(nrows=1, ncols=3 ,figsize=(15,10))

# Gráfico do Total de Casos
sns.barplot(x='location', y=total_casos_data_recente['total_cases'], data=paises_selecionados, ax=ax[0]).set_title('Número Total de Casos')
ax[0].set_ylabel('Total de Casos em escala de 1:10 milhões')
ax[0].set_xlabel('Localização')

# Gráfico do Total de Óbitos
sns.barplot(x='location', y=total_mortes_data_recente['total_deaths'], data=paises_selecionados,ax=ax[1]).set_title('Número Total de Óbitos')
ax[1].set_ylabel('Total de Óbitos')
ax[1].set_xlabel('Localização')

# Gráfico da Densidade Populacional
sns.barplot(x='location', y='population_density', data=paises_selecionados).set_title('Densidade Populacional em Km²')
ax[2].set_ylabel('Densidade')
ax[2].set_xlabel('Localização')

plt.tight_layout();
fig.savefig('total_barras.png')
```

Observando os gráficos obtidos temos alguns destaques importantes:

* A Índia, segundo país mais populoso do mundo, possui menos da metade do número de casos confirmados, se comparados com os Estados Unidos, que tem sua densidade populacional quase 13 vezes menor.

* Comparando ainda a Índia com o Brasil, temos que a densidade populacional também não foi o fator influente para crescimento no número de casos, já que o Brasil possui uma densidade ainda menor que os Estados Unidos e está com os números de casos bastante próximos aos da Índia.

* O México é interessante de se observar, pois possui uma densidade populacional maior que os Estados Unidos e o Brasil, e os seus registros de casos são bastante baixos em relação aos outros países analisados. Entretanto mesmo tendo poucos casos registrados, o número de óbitos é o terceiro na classificação mundial, superando a Índia.

* Os Estados Unidos lidera a classificação do número de casos e óbitos.

Tendo essas informações, podemos ver que a densidade populacional não é o fator decisivo para a disseminação de doenças, o que é estranho de se pensar, pois associamos naturalmente que quanto maior o número de pessoas em um certo espaço, maior seria o descontrole e propiciaria um ambiente ideal para surtos.

Agora que analisamos os gráficos referentes à data mais recente, vamos plotar alguns gráficos em relação ao tempo decorrido desde o início dos registros.

```
# Localizando o primeiro dia de registro de vacinações em cada país selecionado
dia_1_usa = usa.loc[usa.total_vaccinations !=0].sort_values(by='date')[:1]
dia_1_bra = bra.loc[bra.total_vaccinations !=0].sort_values(by='date')[:1]
dia_1_ind = ind.loc[ind.total_vaccinations !=0].sort_values(by='date')[:1]
dia_1_mex = mex.loc[mex.total_vaccinations !=0].sort_values(by='date')[:1]

# Criando um DataFrame para as informações do dia 1
dia_1_paises_selecionados = dia_1_usa.append(dia_1_bra).append(dia_1_ind).append(dia_1_mex)
dia_1_paises_selecionados.sort_values(by='date')
```

Agora partiremos para a análise do número total de vacinações e a média de aplicações diárias desde o início das campanhas.

```
# Localizando a data em que começaram as vacinações para ajustar a escala do gráfico do total de vacinações
df_data_clean_recente = df_data_clean.copy()
df_data_clean_recente = df_data_clean_recente.loc[df_data_clean.total_vaccinations != 0]

# Calculando a média dos registros de novas vacinações para cada país 
media_vacinacao_usa = paises_selecionados.loc[paises_selecionados.iso_code == 'USA','new_vaccinations'].mean()
media_vacinacao_bra = paises_selecionados.loc[paises_selecionados.iso_code == 'BRA','new_vaccinations'].mean()
media_vacinacao_ind = paises_selecionados.loc[paises_selecionados.iso_code == 'IND','new_vaccinations'].mean()
media_vacinacao_mex = paises_selecionados.loc[paises_selecionados.iso_code == 'MEX','new_vaccinations'].mean()

# Criando um novo DataFrame para as médias
df_media_vacinacao_diaria = {
    'location':['United States','Brazil','India','Mexico'],
    'mean_new_vaccinations':[media_vacinacao_usa,media_vacinacao_bra,media_vacinacao_ind,media_vacinacao_mex]
    }
df_media_vacinacao_diaria = pd.DataFrame(df_media_vacinacao_diaria)

# Criando o Plot
sns.set_theme()
fig, ax = plt.subplots(nrows=2, ncols=1 ,figsize=(15,10))

# Plotando o gráfico do total de vacinações desde o início das campanhas de vacinações
sns.lineplot(data=df_data_clean_recente, x=df_data_clean_recente['date'],y=df_data_clean_recente['total_vaccinations'], hue=paises_selecionados['location'], ax=ax[0]).set_title('Total de Vacinações em Relação ao Tempo')
ax[0].set_xlabel('Data')
ax[0].set_ylabel('Total de Vacinações em escala de 1:100 milhões')

# Plotando o gráfico para as médias de vacinações diárias 
media_vacinacao = df_media_vacinacao_diaria.sort_values(by="mean_new_vaccinations", ascending=False)

sns.barplot(x=media_vacinacao['location'], y=media_vacinacao['mean_new_vaccinations'], data= df_data_clean_recente, ax=ax[1]).set_title('Média de Vacinas Administradas Diariamente')
ax[1].set_ylabel('Média de Vacinação')
ax[1].set_xlabel('Localização');

plt.tight_layout();
fig.savefig('media_vacinas.png')
```

Olhando o gráfico e as informações obtidas anteriormente é possível reparar dois países totalmente opostos, os Estados Unidos e o México, em que mesmo tendo iniciado os registros de vacinações na mesma semana, o México acabou ficando para atrás do Brasil e a Índia, sendo que estes dois últimos só iniciaram os registros quase um mês depois. 

O México possui uma média diária de 28,2 mil doses administradas, já os Estados Unidos, que disparou na frente nas campanhas, possui uma média diária de 434,5 mil vacinações.

Outro caso semelhante é a relação entre a Índia e o Brasil, que também tendo começado os registros praticamente ao mesmo tempo, diferença de apenas um dia, o Brasil também ficou para trás, sendo a Índia com uma média diária de 277,2 mil doses e o Brasil 69,9 mil.

Esses resultados podem ser derivados de inúmeros fatores, como o poder de compra de cada país, investimentos dos governos contra o vírus, estratégias de vacinações bem estabelecidas, aprovação da população quanto a segurança das vacinas, entre outros.
